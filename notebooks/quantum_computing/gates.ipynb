# Introduction

A quantum gate is a linear operator (matrix) that evolves a quantum state $|\psi\rangle$ in a given basis. Given an arbitrary quantum state $|\psi\rangle$ and a gate $U$, the action of the gate upon such statevector is represented by:

$$U|\psi\rangle,$$

which is a product between the matrix representing the gate and the vector representing the quantum state.

When a given state $|\psi_j\rangle$ is said to be an eigenstate of the quantum gate (linear operator), this action is described by an eigenvalue-eigenvector equation:

$$U |\psi_j\rangle = \lambda_j |\psi_j\rangle.$$
