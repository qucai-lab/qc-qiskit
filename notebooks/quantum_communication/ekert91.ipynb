# Theory

Entanglement-based protocols rely on the monogamy of entanglement to ensure that if Alice and Bob share strong quantum correlations, no third party (eavesdropper) can be entangled with them in the same way, i.e, being able to share the same correlations without being detected. Security is also supported by the no-cloning theorem.
