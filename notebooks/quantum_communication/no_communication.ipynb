# Theorem

The holistic feature of entangled systems is constrained by the `no-communication theorem`, which states that:

_Entanglement or shared quantum states, alone, cannot be used for transmission of classical information (bits)._

Corollary: _Quantum correlations cannot be used for faster-than-light (superluminal) communication._

Question: Does superdense coding violate the no-communication theorem?

Answer: No, because Alice still needs to physically transmit her qubit (the particle) to Bob through a quantum channel after applying her unitary operation. The transmission of the qubit is essential for Bob to perform a Bell-state measurement and decode the two classical bits. Without this step, Bob would not be able to retrieve any information about Alice's message, thus preserving the no-communication theorem.
