# Projection noise

Projection noise is noise due to `sampling`, i.e., the probability of observing an eigenstate of a superposition state changes with the `number of measurements` (shots). The probability tends to the theoretical value as the number of shots goes to infinity.

# Measurement noise

Measurement noise a.k.a State Preparation And Measurement (SPAM) error. Is the `misclassification` of measuring state $|0\rangle$ when state $|1\rangle$ was prepared. **Mitigation:** One can then multiply the inverse of the confusion matrix by the noisy probabilities to mitigate the results of the noisy experiment: 

\begin{equation}
p_{ideal} = A^{-1} \ p_{noisy}.
\end{equation}

# Coherent noise

Coherent noise is the result of `miscalibration`, when a rotation by $\theta$ turns into $\theta+\epsilon$.

# Incoherent noise

Incoherent noise is the noise that arises from `decoherence` or `imperfect control`. Examples: depolarizing noise channel.

**Note:** Recall that coherent noise is described by unitary operations, while incoherent noise generates mixed states that are always random, regardless of the basis in which they are measured.
