# Definition

Entanglement is a special kind of quantum superposition and it cannot be created from a separable state via [Local Operations and Classical Communication (LOCC)](https://en.wikipedia.org/wiki/LOCC) acting on each of its subsystems, it can either be preserved or destroyed. 

Quantum entanglement paved the way for the realization of many quantum processes in quantum communication, such as `Quantum Teleportation (QT)` and `Superdense Coding (SC)`. Many of which have been realized by optical experiments, in particular, SC that has been reported using parametric down-conversion.
