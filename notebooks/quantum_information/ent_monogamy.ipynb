# Definition

Entanglement monogamy states that if two quantum systems, $A$ and $B$, are maximally entangled, then neither can share entanglement with a third system, $C$.

# Does the GHZ state violate monogamy?

The Greenberger-Horne-Zeilinger (GHZ) state is a maximally entangled state of three qubits defined as:

$$|GHZ\rangle^{3qb}_{ABC} \doteq \frac{1}{\sqrt 2} \left ( |000\rangle_{ABC} +  |111\rangle_{ABC} \right ).$$

It does not violate monogamy because:
- The entanglement involves all three qubits simultaneously rather than pairwise entanglement between two qubits.
- After tracing out one of the qubits (e.g., qubit _C_), the remaining two qubits (_A_ and _B_) would be in a classical mixture of two product states, i.e., a separable (non-entangled) state:

$$ \rho_{ab} = Tr_c(\rho_{abc}) =  Tr_c(|GHZ\rangle \langle GHZ|) = \frac{1}{2} |00\rangle \langle 00| + |11\rangle \langle 11|.$$
