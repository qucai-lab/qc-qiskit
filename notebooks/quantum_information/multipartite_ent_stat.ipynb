# Introduction

An entangled state is a special kind of quantum superposition state, so it is reasonable to write the global state vector of an entangled system using the superposition principle. 

Let $\mathcal{H}_{d}=\otimes_{s=1}^N \mathcal{H}_s^{d_s}=\mathcal{H}_1^{d_1} \otimes \mathcal{H}_2^{d_2} \otimes \cdots \otimes \mathcal{H}_N^{d_N}$ denote the composite Hilbert space of a multipartite quantum system with $N$ qudits, where the qudit of subsystem $s$ is written in an orthonormal basis set $\{|j_s\rangle \}_{j_s=1}^{d_s}$. 

From the superposition principle, the $N$-qudit `global state vector` of this multipartite entangled system has the form:

\begin{align}
|\psi_{1\cdots N}^{entg}\rangle_{Nqd} = \sum_{j_1,\cdots,j_N=1}^{d_1\cdots d_N} c_{j_1\cdots j_N} |j_1\rangle_{1} \otimes \cdots \otimes |j_N\rangle_{N}.
\end{align}

# Separable vs Entangled States

In general, a bipartite quantum system represented by a **pure state vector** $|\psi\rangle_{ab} \in \mathcal{H}_{ab}$ is entangled if and only if it is not a product state ([separable state](https://en.wikipedia.org/wiki/Separable_state)) of its subsystems, i.e, one cannot write a **entangled pure state** as a tensor product of the state vectors of its subsystems.

- A **pure state** is entangled if and only if it cannot be written as a tensor product of pure states:

$$|\psi^{entg}\rangle_{ab} \neq  |\psi^{sep}\rangle_{ab} = |\psi\rangle_{a} \otimes |\psi\rangle_{b}.$$

- A **mixed state** is entangled if and only if it cannot be written as a convex sum of product states:

$$\rho_{ab}^{entg} \neq  \rho_{ab}^{sep} = \sum_i p_i \rho_{a}^{(i)} \otimes \rho_{b}^{(i)}, \text{ with} \sum_i p_i = 1.$$



Consider the particular case of a bipartite quantum system in the pure state $|\psi\rangle_{ab}^{2qd} \in$ $\mathcal{H}_a \otimes \mathcal{H}_b$ of a composite Hilbert space with some set of orthonormal basis $\{|i\rangle\}_{i=1}^{d_a}$ $\in \mathcal{H}_a$ of subsystem $a$, and $\{|j\rangle\}_{j=1}^{d_b}$ $\in \mathcal{H}_b$ of subsystem $b$. One can always write the bipartite global state in the form:

\begin{equation}
|\psi\rangle_{ab}^{2qd} = \sum_{i=0}^{d_a -1}\sum_{j=0}^{d_b -1} c_{i j} |i\rangle_a \otimes |j\rangle_b,
\end{equation}

with $c_{i j}= \langle i | \otimes \langle j |\psi_{ab}\rangle$.

The above global state is separable if and only if the state of its subsystems can be written as state vectors of the form:

\begin{equation}
|\psi\rangle_a^{1qd} = \sum_{i=1}^{d_a} c_{i} |i\rangle_{a},
\end{equation}

and 

\begin{equation}
|\psi\rangle_b^{1qd} = \sum_{j=1}^{d_b} c_{j} |j\rangle_{b},
\end{equation} 

with $c_{i j}=c_{i} c_{j}$. 

Otherwise, the global state is entangled (not separable). For an entangled pure state, each subsystem is described by a mixed reduced density operator, not by a pure state vector.

Any attempt to write an entangled state as a tensor product of the states of its subsystems yields an absurd. Let us demonstrate this by equating the Bell state $|\phi^{Bell +}\rangle_{AB}$ to a tensor product of state vectors $|\phi\rangle_A \doteq \alpha_1|0\rangle+\beta_1 |1\rangle$ and $|\phi\rangle_B \doteq\alpha_2|0\rangle+\beta_2 |1\rangle$, as follows:

\begin{align}
    |\phi^{sep}\rangle_{AB} &= |\phi\rangle_A \otimes |\phi\rangle_B \\
    &=\alpha_1\alpha_2 |00\rangle+ \alpha_1\beta_2 |01\rangle \\
    &+ \beta_1\alpha_2 |10\rangle+ \beta_1\beta_2 |11\rangle \\
    &= \frac{1}{\sqrt{2}}\left( |00\rangle_{AB} + |11\rangle_{AB}  \right) \\
    &\implies \alpha_1\alpha_2=\beta_1\beta_2=\frac{1}{\sqrt{2}} \\
    &\implies \alpha_1\beta_2 \wedge \beta_1\alpha_2 \neq 0,
\end{align}

which is absurd since the last coefficients must vanish. Therefore, the global state of the composite system is said to be entangled. 

# Bipartite Maximally Entangled Qudit States

Bipartite maximally entangled qudit states have the form:

\begin{align}
|\psi^{entg} \rangle_{ab}^{1qd} = \frac{1}{\sqrt{d}} \sum_{j=0}^{d-1} |j\rangle_a \otimes |j\rangle_b.
\end{align}


1. Maximally entangled single-qubit state in the computational basis $\{|0\rangle, |1\rangle\}$:

\begin{align}
|\psi^{entg} \rangle_{ab}^{1qb} &= \frac{1}{\sqrt{2}} \sum_{j=0}^{1} |j\rangle_a \otimes |j\rangle_b \\
&= \frac{1}{\sqrt{2}} \Big( (|0\rangle_a \otimes |0\rangle_b) + (|1\rangle_a \otimes |1\rangle_b) \Big)\\
&= \frac{1}{\sqrt{2}} \Big( |00\rangle_{ab} + |11\rangle_{ab} \Big).
\end{align}

1. Maximally entangled single-qutrit state in the computational basis $\{|0\rangle, |1\rangle, |2\rangle\}$:

\begin{align}
|\psi^{entg} \rangle_{ab}^{1qt} &= \frac{1}{\sqrt{3}} \sum_{j=0}^{2} |j\rangle_a \otimes |j\rangle_b \\
&= \frac{1}{\sqrt{3}} \Big( (|0\rangle_a \otimes |0\rangle_b) + (|1\rangle_a \otimes |1\rangle_b) + (|2\rangle_a \otimes |2\rangle_b) \Big)\\
&= \frac{1}{\sqrt{3}} \Big( |00\rangle_{ab} + |11\rangle_{ab} + |22\rangle_{ab} \Big).
\end{align}

# Pure Entangled States with Maximally Mixed Subsystems

One way to quantify the amount of noise (classical ignorance) of a quantum system and, therefore, infer its degree of entanglement, is looking at the purity of the reduced density operator of the global state of that system. Let $\rho$ denote a generic density operator, the Purity of $\rho$ is defined as:

\begin{align}
    \mathcal{P}(\rho)\doteq tr(\rho^{\dagger}\rho)=tr(\rho^2) \leq 1,
\end{align}

with equality if and only if $\rho$ is a pure density operator. In addition, there is a lower bound defined by $tr(\rho^2)\geq\frac{1}{2}$ for which $\rho$ is a maximally mixed state. 

Note: Global states that are completely known are commonly referred to as pure states (whether entangled or not).

- Mixed density operator:
  
\begin{align}
tr(\rho^{\dagger}\rho)=tr(\rho^2) < 1,
\end{align}
and $\rho^2\neq\rho$.

- Pure density operator:
  
\begin{align}
tr(\rho^{\dagger}\rho)=tr(\rho^2)=1,
\end{align}    

and $\rho^2=\rho$ (idempotent).

Consider, for instance, a bipartite quantum system whose global state is the singlet state with associated bipartite pure density operator $\rho_{ab}$. One can show that each subsystem of $\rho_{ab}$ is in a maximally mixed state. To demonstrate that, let us derive the reduced density operator $\rho_a$ of the first qubit in subsystem $a$, as follows:

\begin{align}
\rho_a &= tr_b(\rho_{ab}) =tr_b\left(|\psi_{ab}^{Bell -}\rangle \langle\psi_{ab}^{Bell -}|\right) \\
&=  tr_b \left\{
\left( \frac{|0_a1_b\rangle - |1_a0_b\rangle}{\sqrt{2}} \right)
\left(\frac{\langle0_a1_b| - \langle 1_a0_b|}{\sqrt{2}} \right)
\right\}
\\
&=\frac{1}{2}tr_b
\Bigg\{
|0_a1_b\rangle\langle0_a1_b|-|0_a1_b\rangle\langle1_a0_b|
-|1_a0_b\rangle\langle0_a1_b|+|1_a0_b\rangle\langle1_a0_b|
\Bigg\}\\
&=\frac{1}{2}\Bigg\{tr_b(|0_a\rangle\langle0_a|\otimes|1_b\rangle\langle1_b|
-tr_b(|0_a\rangle\langle1_a|\otimes|1_b\rangle\langle0_b|)
-tr_b(|1_a\rangle\langle0_a|\otimes|0_b\rangle\langle1_b|)
+tr_b(|1_a\rangle\langle1_a|\otimes|0_b\rangle\langle0_b|))\Bigg\}.
\end{align}

And with the partial trace equation, one finally has:

\begin{align}
\rho_a&=\frac{1}{2}
\Bigg\{
|0_a\rangle \langle 0_a| \langle 1_b|1_b\rangle  - |0_a\rangle \langle 1_a| \langle 1_b|0_b\rangle 
- |1_a\rangle \langle 0_a| \langle 0_b|1_b\rangle + |1_a\rangle \langle 1_a| \langle 0_b|0_b\rangle
\Bigg\}\\
&=\frac{|0\rangle \langle 0|+|1\rangle \langle 1|}{2} = \frac{\mathbb{I}_2}{2} = 
\begin{bmatrix} 
0.5 && 0  \\
0 && 0.5  
\end{bmatrix}.
\end{align}

Since $tr(\rho_a^2)=tr(\rho_b^2)=tr((\mathbb{I}_2/2)^2)=\frac{1}{4}tr(\mathbb{I}_2^2)=1/2$, such a subsystem is in a maximally mixed state. 


When the global state of the composite system is prepared in a maximally entangled pure state $|\psi_{ab}\rangle$, i.e, there is no classical ignorance (see density operator formalism) associated to its preparation (is a pure state), the state of any of its subsystems after tracing out (discarding) one of the qubits is in a maximally mixed state. Said mixed state is a state of maximum entropy with the least amount of classical knowledge (predictability power) about the state of the system upon measurement. A mixed state is represented by an `incoherent density matrix` (off-diagonal entries are all zero) in the density operator formalism, where the diagonal entries denote a uniform probability distribution over states $|0\rangle$ and $|1\rangle$.

\begin{equation}
tr_b(\rho_{ab}) = \rho_a \doteq \sum_{j=1}^n  p_j |\psi_j \rangle \langle \psi_j| = \frac{|0\rangle \langle 0|+|1\rangle \langle 1|}{2} = \frac{\mathbb{I}_2}{2} = 
\begin{bmatrix} 
0.5 && 0  \\
0 && 0.5  
\end{bmatrix}.
\end{equation}

In general, any maximally entangled state is necessarily a pure state, and its subsystems are each in a maximally mixed state. However, it is possible to prepare a pure state that is not maximally entangled.
