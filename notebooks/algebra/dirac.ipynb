# Dirac Notation

- A $d$-dimensional **Ket state** $|\psi\rangle$: is a $d \times 1$ tensor (column vector) defined as:

$$|\psi\rangle := \begin{bmatrix} \psi_1 \\ \psi_2 \\ \vdots \\ \psi_d \end{bmatrix}.$$

- A $d$-dimensional **Bra state** $\langle \psi| = (|\psi\rangle)^{\dagger}$: a.k.a dual vector (transpose conjugate) is a $1 \times d$ tensor (row vector) defined as:

$$\langle \psi| := (|\psi\rangle)^{\dagger}=(|\psi\rangle^T)^*  = \begin{bmatrix} \psi_1^{*}& \psi_2^{*} \cdots \psi^{*}_d \end{bmatrix}.$$
