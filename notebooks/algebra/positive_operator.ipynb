# Definition



Positive operators are a special subclass of Hermitian operators. A Positive operator $\hat{J}$ is defined as a linear operator such that, for any vector $|\psi \rangle$, the inner product $(|\psi \rangle, \hat{J}|\psi \rangle)$ is a real and non-negative number. If the inner product $(|\psi \rangle, \hat{J}|\psi \rangle)$ is strictly greater than zero for all $|\psi \rangle \neq 0$ then $\hat{J}$ is termed a positive definite operator. Within Dirac's bra-ket notation, a positive definite operator $\hat{J}$ reads:

\begin{equation}
\langle \psi |\hat{J}| \psi \rangle > 0,
\end{equation}

for all $|\psi\rangle \neq |\emptyset\rangle$.

While a positive semidefinite operator $\hat{J}$ reads:

\begin{equation}
\langle \psi |\hat{J}| \psi \rangle \geq 0,
\end{equation}

for all $|\psi\rangle$.

# Properties



1. The eigenvalues of a positive operator are all non-negatives.

\begin{align}
\langle \phi |\hat{\mathcal{O}}| \phi \rangle &= \sum_{j,k} c_j^*c_k \langle \phi_j |\hat{\mathcal{O}} | \phi_k \rangle \\ &=\sum_{j,k} c_j^*c_k \langle \phi_j | o_k |\phi_k \rangle \\
&= \sum_{j,k} c_j^*c_k o_k \langle \phi_j |\phi_k \rangle 
    \\&= \sum_{j,k} c_j^*c_k o_k \delta_{jk} 
    \\&= \sum_{j=1}^2 |c_j|^2 o_j \geq 0 \implies o_j \geq 0.
\end{align}

  
2. A positive operator is always Hermitian.

$$ \hat{\mathcal{O}} = \hat{\mathcal{O}}^\dagger. $$
  
3. The positiveness of a density operator can be evaluated by looking at the coefficients of its characteristic polynomial equation.

# Applications


1. In the context of quantum measurements, positive operators are used to define the `Positive Operator-Valued Measures (POVMs)`, which generalize the concept of projective measurements. POVMs allow for a broader range of measurement outcomes and are essential in various quantum information processing tasks, such as quantum state discrimination and quantum cryptography.

2. In the density operator (a.k.a density matrix) formalism used to describe `pure and mixed quantum states`, every density operator is a positive-semidefinite operator with unit trace. Particularly, the unit trace condition ensures that the probabilities derived from the density operator are always non-negative and sum to one, which is a fundamental requirement in probability theory.

3. Another application of positive operators is in the formulation of quantum channels, which describe the evolution of quantum states in open quantum systems. Quantum channels can be represented using `Completely Positive Trace-Preserving (CPTP)` maps, which ensure that the physicality of quantum states is preserved during their evolution.

4. In entanglement theory, positive operators play a crucial role in characterizing entangled states and separability criteria. The [Peres-Horodecki criterion](https://en.wikipedia.org/wiki/Peres%E2%80%93Horodecki_criterion), for instance, uses the concept of `Positive Partial Transpose (PPT)` to determine the separability of bipartite quantum states. The Peresâ€“Horodecki criterion states that if a bipartite quantum state is separable, then its partial transpose is positive semidefinite, i.e, all the eigenvalues of the partial transpose are non-negative. For $2\times2$ and $2\times3$ systems, this condition is sufficient for separability, for higher-dimensional bipartite systems, positivity of the partial transpose does not guarantee separability, because there exist entangled states whose partial transpose is still positive, known as [bound entangled states](https://en.wikipedia.org/wiki/Bound_entanglement).
   - Key Takeaways:
     - All separable states have positive partial transpose (PPT), but some entangled states also have PPT.
     - Separable $\implies$ PPT.
     - PPT $\not\implies$ Separable.
