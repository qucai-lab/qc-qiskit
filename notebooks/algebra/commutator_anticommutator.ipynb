# Definition <a name="definition"></a>

The **commutator** of two operators $\hat{A}$ and $\hat{B}$ is defined as:

$$[\hat{A}, \hat{B}] = \hat{A}\hat{B} - \hat{B}\hat{A}.$$

The **anticommutator** of two operators $\hat{A}$ and $\hat{B}$ is defined as:

$$\{\hat{A}, \hat{B}\} = \hat{A}\hat{B} + \hat{B}\hat{A}.$$

# Properties <a name="properties"></a>

1. **Relation to Identity**:
   - If $\hat{A}$ and $\hat{B}$ commute, then $[\hat{A}, \hat{B}] = 0 \implies \hat{A}\hat{B} = \hat{B}\hat{A}$.
   - If $\hat{A}$ and $\hat{B}$ anticommute, then $\{\hat{A}, \hat{B}\} = 0 \implies \hat{A}\hat{B} = -\hat{B}\hat{A}$.

2. **Symmetry**:
   - Commutator: $[\hat{A}, \hat{B}] = -[\hat{B}, \hat{A}]$.
   - Anticommutator: $\{\hat{A}, \hat{B}\} = \{\hat{B}, \hat{A}\}$.

3. **Linearity**:
   - Commutator: $[a\hat{A} + b\hat{B}, \hat{C}] = a[\hat{A}, \hat{C}] + b[\hat{B}, \hat{C}]$.
   - Anticommutator: $\{a\hat{A} + b\hat{B}, \hat{C}\} = a\{\hat{A}, \hat{C}\} + b\{\hat{B}, \hat{C}\}$.

4. **Jacobi Identity** (for commutators):
   $$[\hat{A}, [\hat{B}, \hat{C}]] + [\hat{B}, [\hat{C}, \hat{A}]] + [\hat{C}, [\hat{A}, \hat{B}]] = 0.$$

# Applications <a name="applications"></a>

1. **Quantum Mechanics**: 
   - Commutators define the uncertainty relations between conjugate observables: $\sigma_A \sigma_B \geq \frac{1}{2} |\langle [\hat{A}, \hat{B}] \rangle|$.
   - The canonical commutation relation $[\hat{x}, \hat{p}] = i\hbar \mathbb{I}_d$ is a hallmark of quantum mechanics and has been widely used to classify the different types of superconducting circuit-based qubits. In particular, for conjugate observables such as electric charge $Q \to \hat{Q}$ and magnetic flux $\Phi \to \hat{\Phi}$, the commutation relation yields the same result as the prime example of position and momentum: $[\hat{\Phi}, \hat{Q}] = \left(\hat{\Phi}\hat{Q}- \hat{Q}\hat{\Phi}\right) = i\hbar \mathbb{I}_d.$
 
2. **Lie Algebras**: Commutators are used to define the structure of Lie algebras, which are essential in the study of symmetries in physics.
