# Hilbert Space


The equivalence between wave mechanics and matrix mechanics is primarily attributed to John Von Neumann, who in 1932 proved them to be different realizations of an abstract Hilbert space.

- Definition: Let $\{|\lambda_j\rangle\}_{j=1}^d$ denote a linearly independent set of orthonormal column vectors $|\lambda_j\rangle$ in a finite dimensional complex vector space $\mathcal{H}$ of dimension $\dim \mathcal{H} = d$ known as Hilbert space. If the set is a spanning set for $\mathcal{H}$, i.e, $span (\{|\lambda_j\rangle\}_{j=1}^d) = \mathcal{H}$, then the set is a basis set and the vector space $\mathcal{H}$ is said to be generated or spanned by the basis set.

The number of linearly independent vectors in the basis set is equal to the dimension $d$ of the space. Moreover, linear independence states that no vector in the set can be given as a linear combination of the remaining ones and, therefore, the set form a basis for the space $V^d$. As a consequence, any arbitrary state vector (a.k.a ket-vector or column vector) $|\psi\rangle \in V^d$ can be represented as a linear combination of the vectors of the basis set according to: 

\begin{equation}
|\psi\rangle \doteq \sum_{j=1}^d \lambda_j|\lambda_j\rangle = \begin{pmatrix} \lambda_1 \\ \lambda_2 \\ \vdots \\ \lambda_d \end{pmatrix}.
\end{equation}

And the transposed conjugate state vector also known as bra-vector or row vector is defined as:

\begin{align}
\langle \psi| = (|\psi\rangle)^{\dagger}=(|\psi\rangle^T)^* 
= \left( \sum_{j=1}^d \lambda_j | \lambda_j \rangle \right)^{\dagger}
= \sum_{j=1}^d \lambda_j^* \langle\lambda_j|
= \begin{pmatrix}\lambda_1^* & \lambda_2^* & \cdots & \lambda_d^* \end{pmatrix}.
\end{align}

Here, $|\psi\rangle^T$ denotes the transpose of $|\psi\rangle$ swapping rows for columns. The complex scalar quantity $\lambda_j$ is defined in the standard way $\lambda_j\doteq a+bi$ with real scalars $a$ and $b$, imaginary number $i\doteq \sqrt{-1}$ and modulus $|\lambda|=\sqrt{\lambda \lambda^*}=\sqrt{a^2+b^2}$. Whereas $(*)$ denotes the complex conjugate: $\lambda_j^*=a-ib$. 
The conjugate transpose (a.k.a Hermitian conjugate, self-adjoint or dagger) denoted by $\dagger$ has the following properties:


- $(| \psi \rangle+| \phi \rangle)^{\dagger}= | \psi \rangle^{\dagger}+| \phi \rangle^{\dagger}.$
- $(\langle \psi | \phi \rangle)^{\dagger} =(\langle \psi | \phi \rangle)^*= \langle \phi | \psi \rangle.$
- $(\lambda_j | \psi \rangle)^{\dagger}= \lambda_j^*| \psi \rangle^{\dagger}.$
- $(| \psi \rangle^{\dagger})^{\dagger}=| \psi \rangle.$

Orthonormality condition of the basis vectors entails:

\begin{equation}
\langle\lambda_j|\lambda_k\rangle \doteq \delta_{jk} = 
\begin{cases}
0, & \text{if } j \ne k, \\
1, & \text{if } j=k, 
\end{cases}
\end{equation}

where $\delta_{jk}$ is termed `Kronecker delta`. In this sense, the basis set $\{|\lambda_j\rangle\}_{j=1}^d$ is termed an orthonormal basis set. In tensorial notation, the $d$x$d$-dimensional unit matrix (a.k.a identity matrix) reads:

\begin{align}
\delta_{\mu\nu}=
    \begin{bmatrix} 
     1 & \dots & 0 \\
    \vdots & \ddots & \vdots \\
    0 & \dots & 1
\end{bmatrix}.
\end{align}
