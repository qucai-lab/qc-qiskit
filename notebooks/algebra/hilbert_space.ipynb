# Definition


A Hilbert space $\mathcal{H}$ is a complete inner-product space, i.e., a vector space endowed with an inner product $\langle \psi| \phi\rangle = |\psi\rangle^{\dagger}| \phi\rangle$ such that the induced norm 

$$||\psi|| := \sqrt{\langle \psi, \psi \rangle},$$

makes it complete (Banach space) because every Cauchy sequence of vectors converges to a well-defined limit that is within the space. 

- An inner product space is a Hilbert space if and only if it is complete with respect to the norm induced by its inner product.

- Equivalently, If an inner product space is complete (and thus a Banach space with respect to its induced norm), then it is a Hilbert space over the field of real $\mathbb{R}$ or complex $\mathbb{C}$ scalars.

# Composite Spaces

Consider $n$ Hilbert spaces, where $\mathcal{H}_1^{d_1}$ has dimension $d_1$, $\mathcal{H}_2^{d_2}$ has dimension $d_2$, etc.

The Hilbert space $\mathcal{H}_d$ of the composite space is given by the tensor product of each constituent space:

$$\mathcal{H}_{d} := \otimes_{s=1}^N \mathcal{H}_{d_s}=\mathcal{H}_{d_1} \otimes \mathcal{H}_{d_2} \otimes \cdots \otimes \mathcal{H}_{d_N}.$$

The dimension of the composite space is given by the product of the dimensions of each constituent space:

$$d := \dim(\mathcal{H}_{d})=\prod_{s=1}^N d_s = d_1 \times d_2 \times \cdots \times d_N.$$

The orthonormal basis of the composite space is given by the tensor product of the basis vectors of each constituent space: 

$$\{\otimes_{s=1}^N |o_{j_s}\rangle_s\}_{j_s=0}^{d_s-1} := |o_{j_1}\rangle_1 \otimes |o_{j_2}\rangle_2 \otimes \cdots \otimes |o_{j_N}\rangle_N.$$
