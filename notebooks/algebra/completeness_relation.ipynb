# Definition

Suppose \{|$j\rangle\}_{j=1}^d$ is any orthonormal basis set for a $d$-dimensional vector space $\mathbb{V}^d$ with an arbitrary vector $|v\rangle= \sum_{j=1}^d v_j|j\rangle$, then

\begin{equation}
\left(\sum_{j=1}^d |j\rangle \langle j| \right) |v\rangle = \sum_{j=1}^d |j\rangle \langle j|v\rangle \\
= \sum_{j=1}^d v_j |j\rangle = |v\rangle.
\end{equation}

And, therefore,

\begin{equation}
\sum_{j=1}^d |j\rangle \langle j| = \mathbb{I}_d,
\end{equation}

known as the completeness relation.


# Applications 

- The completeness relation is widely used in quantum mechanics and quantum computing to express the identity operator in terms of a complete set of orthonormal basis states. This is crucial for expanding quantum states and operators in different bases.
