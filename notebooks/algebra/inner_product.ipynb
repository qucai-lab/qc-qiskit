# Inner Product

## Definition

Every vector space is equipped with an operation called the inner product (or dot product).

A function named inner product is a positive semi-definite bilinear map defined as $\langle \cdot,\cdot \rangle : \mathbb{V}^d$ x $\mathbb{V}^d \rightarrow \mathbb{C}$. The map is voiced: "a function $\langle \cdot,\cdot \rangle$, namely inner product, takes as input two vectors from a vector space and produces a scalar quantity as output". Every inner product induces a norm by: 

\begin{equation} 
|| \mathbb{V} || = \sqrt{\langle \mathbb{V}, \mathbb{V} \rangle}. 
\end{equation}

In this sense, every inner product space is a normed vector space. If the normed space is also complete (a Banach space), the inner product space is a Hilbert space.

In a finite dimensional Euclidean space, the inner product between coordinate vectors $\vec{a}$ and $\vec{b}$ is the dot product:

$$ \vec{a} \cdot \vec{b} = \sum_{i, j=1}^d a_i b_j \delta_{ij} = \sum_{j=1}^d a_j b_j = a_1 b_1 + a_2b_2 + \cdots + a_db_d .$$

## Properties

1. $\langle v|A w\rangle \equiv (|v\rangle, A|w\rangle)=(A^{\dagger}|v\rangle,|w\rangle).$ Is the inner product for the *adjoint* (a.k.a *Hermitian conjugate*) of a linear operator $A$ on a Hilbert space $\mathbb{C}^d$.

## Applications

1. In Quantum Mechanics, the inner product is used to calculate probabilities and expectation values of observables.

# Vector representation for the inner product on a Hilbert Space

Let $|v\rangle = \sum_{i=1}^{\dim \mathbb{V}} v_i |i \rangle$ and $|w\rangle = \sum_{j=1}^{\dim \mathbb{W}} w_i |j \rangle$ be complex vectors with respect to some orthonormal basis on a Hilbert space $\mathbb{C}^d$. Their inner product can be defined as:

\begin{equation}
\langle v | w \rangle \equiv (|v\rangle, |w\rangle) &\doteq& |v\rangle^{\dagger} |w\rangle = \left( \sum_{i=1}^{\dim \mathbb{V}} v_i |i \rangle \right)^{\dagger} \left(\sum_{j=1}^{\dim \mathbb{W}} w_j | j \rangle\right) = \sum_{i=1}^{\dim \mathbb{V}} \sum_{j=1}^{\dim \mathbb{W}}  v_i^*w_j \langle i|j\rangle = \sum_{i,j=1}^d v_i^* w_j \delta_{ij} \\
&=&
\sum_{i=1}^d v_i^* w_i = v_1^* w_1 + v_2^* w_2 + \cdots + v_d^* w_d \\
&=& 
\begin{pmatrix} v_1^* & v_2^* & \cdots & v_d^* \end{pmatrix}
\begin{pmatrix} w_1 \\ w_2 \\ \vdots \\ w_d \end{pmatrix}.
\end{equation}

- The result of the inner product $\langle v | w \rangle$ is a complex scalar value.

**Note:** $\mathbb{C}^d$ is used interchangeably to denote the field of complex scalars and a complex vector space.



# Outer Product 

## Definition

In the outer product representation of Linear Operators, an arbitrary operator $\hat{O}$ can be expressed as a sum of outer products of basis vectors scaled by their corresponding eigenvalues:

$$
\begin{equation}
\hat{O} = I_v \hat{O} I_w = \sum_{j=1}^{d_a} | v_j \rangle \langle v_j| \hat{O} \sum_{k=1}^{d_b}|w_k\rangle \langle w_k|
\\
= \sum_{jk=1}^{d_a,d_b} \langle v_j |\hat{O} |w_k\rangle |v_j\rangle \langle w_k|,
\end{equation} 
$$

- The result of the outer product is a linear operator that maps vectors from one vector space to another.

Equivalent representation:
$$ \hat{A} = \sum_{i} \lambda_i |u_i\rangle \langle u_i|.$$

Where:
- $\lambda_i$ are the eigenvalues of the operator $\hat{A}$.
- $|u_i\rangle$ are the eigenvectors of the operator $\hat{A}$.
- $\langle u_i|$ is the conjugate transpose (bra) of the eigenvector $|u_i\rangle$ (ket).

## Applications

1. In linear algebra, the outer product can be used to change the basis of a vector space or to construct new vector spaces from existing ones.
   
2. In quantum mechanics, the outer product is used to represent operators, such as projection operators and density matrices.

3. It is also used in quantum computing to describe quantum gates and transformations on quantum states.

## Examples

The Pauli matrices can then be rewritten in terms of outer products as:
    
\begin{align}
\hat{X} &= \sum_{j,k=0}^{1} \langle j |\sigma_1|k\rangle |j\rangle \langle k| = |0\rangle\langle 1|+|1\rangle\langle 0|.\\
\hat{Y} &= \sum_{j,k=0}^{1} \langle j |\sigma_2|k\rangle |j\rangle \langle k|=-i|0\rangle\langle 1|+i|1\rangle\langle 0|.\\
\hat{Z} &= \sum_{j,k=0}^{1} \langle j |\sigma_3|k\rangle |j\rangle \langle k|=|0\rangle\langle 0|-|1\rangle\langle 1|.\\
\end{align}

And also:

$$ \hat{X} = +1 |+\rangle \langle +| - 1 |-\rangle \langle -| $$

```python
import numpy as np

zero_1d=np.array([1,0])
one_1d=np.array([0,1])

zero_1d.shape
```

```python
def outer_product_repr(zero, one):
    '''
    When using np.outer(), the input vectors are flattened if not already 1-dimensional.
    '''
    print(f'X gate:\n\n {np.outer(zero,one)+np.outer(one,zero)}\n')
    print(f'Y gate:\n\n {-1j*np.outer(zero,one)+1j*np.outer(one,zero)}\n')
    print(f'Z gate:\n\n {np.outer(zero,zero)-np.outer(one,one)}\n')
    
outer_product_repr(zero_1d, one_1d)
```
