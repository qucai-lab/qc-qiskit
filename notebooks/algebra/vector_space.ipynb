# Definition


A vector space $\mathbb{V}$ over a field $\mathbb{F}$ is a set $\{|v\rangle\}$ of objects called vectors where the operations of **vector addition** and **scalar multiplication** are well defined and satisfy certain requirements, called axioms, listed below.

For vector addition ($+ : \mathbb{V} \times \mathbb{V} \rightarrow \mathbb{V}$):

1. **Closure under addition**: $\forall |u\rangle, |v\rangle \in \mathbb{V}$, the sum $|u\rangle + |v\rangle$ is also in $\mathbb{V}$.
2. **Closure under scalar multiplication**: $\forall a \in F$ and $|v\rangle \in \mathbb{V}$, the product $a|v\rangle$ is also in $\mathbb{V}$.
3. **Commutativity of addition**: $\forall |u\rangle, |v\rangle \in \mathbb{V}$, $|u\rangle + |v\rangle = |v\rangle + |u\rangle$.
4. **Associativity of addition**: $\forall |u\rangle, |v\rangle, |w\rangle \in \mathbb{V}$, $(|u\rangle + |v\rangle) + |w\rangle = |u\rangle + (|v\rangle + |w\rangle)$.
5. **Existence of additive identity**: There exists a vector $|0\rangle \in \mathbb{V}$ such that $\forall |v\rangle \in \mathbb{V}$, $|v\rangle + |0\rangle = |v\rangle$.
6. **Existence of additive inverses**: $\forall |v\rangle \in \mathbb{V}$, there exists a vector $|-v\rangle \in \mathbb{V}$ such that $|v\rangle + |-v\rangle = |0\rangle$.

For scalar multiplication ($* : \mathbb{F} \times \mathbb{V} \rightarrow \mathbb{V}$):

7. **Distributivity of scalar multiplication with respect to vector addition**: $\forall a \in \mathbb{F}$ and $|u\rangle, |v\rangle \in \mathbb{V}$, $a*(|u\rangle + |v\rangle) = a*|u\rangle + a*|v\rangle$.
8. **Distributivity of scalar multiplication with respect to field addition**: $\forall a, b \in \mathbb{F}$ and $|v\rangle \in \mathbb{V}$, $(a + b)*|v\rangle = a*|v\rangle + b*|v\rangle$.
9.  **Associativity of scalar multiplication**: $\forall a, b \in \mathbb{F}$ and $|v\rangle \in \mathbb{V}$, $a*(b*|v\rangle) = (a*b)*|v\rangle$.
10. **Existence of multiplicative identity**: $\forall |v\rangle \in \mathbb{V}$, $\mathbb{1}_{F}*|v\rangle = |v\rangle$, where $\mathbb{1}_{F}$ is the multiplicative identity in the scalar field $\mathbb{F}$. 

Where $\forall$ is voiced "for all".

# Linear Independence

Let $\{|\lambda_j\rangle\}_{j=1}^d$ denote a linearly independent set of orthonormal column vectors $|\lambda_j\rangle$ in a finite dimensional complex vector space $\mathbb{C}^d$ of dimension $\dim \mathbb{C} = d$. If the set is a spanning set for $\mathbb{C}^d$, i.e, $span (\{|\lambda_j\rangle\}_{j=1}^d) = \mathbb{C}^d$, then the set is a basis set and the vector space $\mathbb{C}^d$ is said to be generated or spanned by the basis set.

Orthonormality condition of the basis vectors entails:

\begin{equation}
\langle\lambda_j|\lambda_k\rangle := \delta_{jk} = 
\begin{cases}
0, & \text{if } j \ne k, \\
1, & \text{if } j=k. 
\end{cases}
\end{equation}

Where $\delta_{jk}$ is termed `Kronecker delta`. In tensorial notation, the $d \times d$-dimensional unit matrix (a.k.a identity matrix) reads:

\begin{align}
\delta_{\mu\nu} :=
    \begin{bmatrix} 
     1 & \dots & 0 \\
    \vdots & \ddots & \vdots \\
    0 & \dots & 1
\end{bmatrix}.
\end{align}

The number of linearly independent vectors in the basis set is equal to the dimension $d$ of the space. Moreover, linear independence states that no vector in the set can be given as a linear combination of the remaining ones and, therefore, the set form a basis for the space $\mathbb{C}^d$. 

As a consequence, any arbitrary state vector (a.k.a ket-vector or column vector) $|\psi\rangle \in \mathbb{C}^d$ can be represented as a linear combination of the vectors of the basis set according to: 

\begin{equation}
|\psi\rangle := \sum_{j=1}^d \lambda_j|\lambda_j\rangle = \begin{pmatrix} \lambda_1 \\ \lambda_2 \\ \vdots \\ \lambda_d \end{pmatrix}.
\end{equation}

And the transposed conjugate state vector also known as bra-vector or row vector is defined as:

\begin{align}
\langle \psi|:= (|\psi\rangle)^{\dagger}=(|\psi\rangle^T)^* 
= \left( \sum_{j=1}^d \lambda_j | \lambda_j \rangle \right)^{\dagger}
= \sum_{j=1}^d \lambda_j^* \langle\lambda_j|
= \begin{pmatrix}\lambda_1^* & \lambda_2^* & \cdots & \lambda_d^* \end{pmatrix}.
\end{align}

Here, $|\psi\rangle^T$ denotes the transpose of $|\psi\rangle$ swapping rows for columns. The complex scalar quantity $\lambda_j$ is defined in the standard way $\lambda_j := r_1+ir_2$ with real scalars $r_1$ and $r_2$, imaginary number $i:= \sqrt{-1}$ and modulus $|\lambda| := \sqrt{\lambda \lambda^*} = \sqrt{r_1^2 + r_2^2}$. Whereas $(*)$ denotes the complex conjugate: $\lambda_j^* := r_1 - ir_2$. 

The conjugate transpose (a.k.a Hermitian conjugate, self-adjoint) denoted by $\dagger$ (voiced dagger) has the following properties:


- $(| \psi \rangle+| \phi \rangle)^{\dagger}= | \psi \rangle^{\dagger}+| \phi \rangle^{\dagger}.$
- $(\langle \psi | \phi \rangle)^{\dagger} =(\langle \psi | \phi \rangle)^*= \langle \phi | \psi \rangle.$
- $(\lambda_j | \psi \rangle)^{\dagger}= \lambda_j^*| \psi \rangle^{\dagger}.$
- $(| \psi \rangle^{\dagger})^{\dagger}=| \psi \rangle.$
