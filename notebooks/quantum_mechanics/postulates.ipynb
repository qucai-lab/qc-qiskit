# Introduction

There is no single universally agreed number of postulates in quantum mechanics. The number varies depending on the textbook or author's formulation.

# State Postulate

## State Vector

Any **isolated (close)** quantum system is completely described by a normalized *state vector* $|\psi\rangle$ which is a unit vector embedded in a complex vector space ([Hilbert space](../algebra/hilbert_space.ipynb)) $\mathcal{H}$ known as the *state space* of the system endowed with an inner product $(\cdot,\cdot) := \langle \cdot | \cdot \rangle$.


- **First case: Discrete Spectrum.**

When the spectrum of the eigenvalues is discrete, the state $|\psi\rangle$ of the quantum system is embedded in a finite-dimensional Hilbert space ($|\psi\rangle \in \mathcal{H}^d$) of dimension $d=\dim \mathcal{H}^d$ whose orthonormal basis $\{|o_j\rangle\}_{j=1}^d$ is formed by a finite set of orthonormal vectors. The state of the quantum system in this basis reads:

\begin{equation}
|\psi\rangle := \sum_{j=1}^d c_j|o_j\rangle.
\end{equation}

Since the basis is orthonormal, the inner product becomes:

\begin{equation}
\langle o_j|o_k\rangle := \delta_{jk} = \begin{cases}
0, & \text{if } j \ne k, \\
1, & \text{if } j=k. 
\end{cases}
\end{equation}

With the identity operator in $\mathcal{H}^d$ given by: $$\sum_{j=1}^{d} |o_j \rangle \langle o_j| = \mathbb{I}.$$ 



The normalization condition for the state vector is:

\begin{align}
\langle \psi |\psi\rangle &= |\psi\rangle^{\dagger} |\psi\rangle \\
&= \Bigg(\sum_{j=1}^{d}c_j|o_j\rangle \Bigg)^{\dagger}\Bigg(\sum_{k=1}^{d} c_k |o_k\rangle \Bigg) \\
&= \sum_{j,k=1}^{d} c_j^* c_k \langle o_j|o_k\rangle \\
&= \sum_{j,k=1}^{d} c_j^* c_k \delta_{jk}\\
&= \sum_{j=1}^{d}|c_j|^2 = 1
\end{align}

And probability amplitude $c_j$ associated with the eigenstate $|o_j\rangle$ is obtained via the Fourier trick:

\begin{equation}
c_j := \langle o_j|\psi\rangle.
\end{equation}

- **Second case: Continuous Spectrum.**

When the spectrum of the eigenvalues is continous, the state $|\psi\rangle$ of the quantum system is embedded in a infinite-dimensional Hilbert space ($|\psi\rangle \in \mathcal{H}$) whose orthonormal basis $\{|q\rangle \}$ is formed by an uncountably infinite set of orthonormal vectors. The state of the quantum system in this basis reads:

\begin{equation}
|\psi\rangle := \int \psi_{q} |q\rangle d_{q},
\end{equation}

which is a linear combination given in terms of an integral, rather than a sum (See Ref. [2], Section 3.2.3, pg. 103).

The inner product thus becomes the Dirac delta distribution:

\begin{equation}
\langle q |q'\rangle := \delta (q-q').
\end{equation}

With the identity operator in $\mathcal{H}^d$ given by: $$\int |q \rangle \langle q| d_{q} = \mathbb{I}.$$

The normalization condition for the state vector is:

\begin{equation}
\langle \psi|\psi\rangle = \int |\psi_{q}|^2 d_{q} = 1.
\end{equation}

And the analogous Fourier trick gives the complex-valued scalar wave function $\psi_{q}$ as the probability amplitude:

\begin{equation}
\psi_{q} := \langle q |\psi \rangle.
\end{equation}

Note that $\psi_q$ is the standard wave function of the ubiquitous Schr√∂dinger equation of wave mechanics, often denote by $\psi(q)$ or $\psi(x)$ if in the position space.

## Composite Systems

Consider a quantum system composed of $N$ arbitrary subsystems, where subsystem $s$ is prepared in a quantum state $|\psi_{s}\rangle = \sum_{j=1}^{d_s} c_{j_s} |o_{j}\rangle_s$ embedded in a $d_s$-dimensional Hilbert space $\mathcal{H}_{d_s}$ with orthonormal basis $\{|o_{j}\rangle_s \}|_{j=1}^{d_s}$.

The Hilbert state space $\mathcal{H}_d$ of the composite system is given by the tensor product of each subsystem's state space:

$$\mathcal{H}_{d} := \otimes_{s=1}^N \mathcal{H}_{d_s}=\mathcal{H}_{d_1} \otimes \mathcal{H}_{d_2} \otimes \cdots \otimes \mathcal{H}_{d_N}.$$

The dimension of the composite system's state space is given by the product of the dimensions of each subsystem's state space:

$$d := \dim(\mathcal{H}_{d})=\prod_{s=1}^N d_s = d_1 \times d_2 \times \cdots \times d_N.$$


The orthonormal basis of the composite system is given by the tensor product of the basis vectors of each subsystem: 

$$\{\otimes_{s=1}^N |o_{j_s}\rangle_s\}_{j_s=0}^{d_s-1} := |o_{j_1}\rangle_1 \otimes |o_{j_2}\rangle_2 \otimes \cdots \otimes |o_{j_N}\rangle_N.$$

# Observable Postulate

**Physical observables** (a.k.a **von Neumann observables**) in quantum mechanics are represented by [Hermitian](../algebra/hermitian_operator.ipynb) ($\hat{\mathcal{O}}=\hat{\mathcal{O}}^{\dagger}$) operators, i.e., self-adjoint matrices with real eigenvalues in a $d$-dimensional hilbert space $\mathcal{H}$. Because Hermitian operators are also Normal ($ \hat{\mathcal{O}}\hat{\mathcal{O}}^{\dagger}=\hat{\mathcal{O}}^{\dagger}\hat{\mathcal{O}}$) operators, they have the following [spectral decomposition](../algebra/spectral_decomposition.ipynb) in terms of the outer product representation:

\begin{equation}
\hat{O} = \sum_{j=1}^{d_{\hat{O}}} o_j \Pi_{o_j} = \sum_{j=1}^{d_{\hat{O}}} o_j |o_j \rangle \langle o_j|.
\end{equation}

Where:

- $o_j \in \mathbb{R}$ is the corresponding real eigenvalue of the von Neumann observable $\hat{O}$. 

- $\Pi_{o_j} = |o_j \rangle \langle o_j|$ is the corresponding projector operator of the von Neumann observable $\hat{O}$.

- $\{o_j\}|_{j=1}^d$ is an orthonormal basis set of $d$ linearly independent eigenvectors $|o_j\rangle$ of $\hat{O}$ with eigenvalue $o_j$ satisfying $\hat{O}|o_j \rangle = o_j |o_j \rangle$ and $\langle o_j|o_k \rangle = \delta_{jk}$.

The projector operators satisfy:

\begin{align}
\sum_{j=1}^{d_{\hat{O}}} \Pi_{o_j} = \sum_{j=1}^{d_{\hat{O}}} |o_j \rangle \langle o_j| = \mathbb{I}_{\hat{O}},
\end{align}

\begin{align}
\Pi_{o_j}\Pi_{o_k} = \delta_{jk}\Pi_{o_j}.
\end{align}

# Time Evolution Postulate

The evolution of a **closed quantum system** is described by a *unitary transformation* given by the action of the propagator $\hat{U}_t := \hat{U}(t)$ known as the Unitary time evolution operator. Therefore, the global state $|\psi_t\rangle$ of the system at time $t$ evolves from the initial state $|\psi_0\rangle$ at time $t=0$ according to:

\begin{equation} 
|\psi_t\rangle = \hat{U}_t|\psi_0\rangle.
\end{equation}  

Where:

- $\hat{U}_t^{\dagger} \hat{U}_t = \hat{U}_t \hat{U}_t^{\dagger} = \mathbb{I}$.
- $U_0 = \mathbb{I}$.
- $ |\psi_t\rangle := |\psi (t)\rangle$.
- $ |\psi_0\rangle := |\psi (t_0)\rangle$.

**Note:** In the particular case of closed systems, the evolution operator $\hat{U}_t$ is unitary. However, the evolution operator in open quantum systems is, in general, represented by a non-Unitary operator.

# Measurement Postulate

## General Description for Measurements 

Measurements of a quantum system are described by a collection of *measurement operators* \{$M_m$\} acting on the state space of the system with $m$ possible measurement outcomes. If the quantum system is prepared in a general state $|\psi\rangle$, the probability associated with a measurement outcome $o_m$ is:

\begin{equation} 
Pr(o_m) := \langle \psi | M^{\dagger}_{m} M_{m} |\psi \rangle,
\end{equation} 

and the state of the system immediately after the measurement of the eigenvalue $o_m$ will be:

\begin{align} 
    |\psi_{o_m} \rangle= \frac{M_{m} |\psi \rangle}{\sqrt{\langle \psi | M^{\dagger}_{m} M_{m} |\psi \rangle}}. 
\end{align}

The measurement operators satisfy the completeness relation:

\begin{equation} 
\sum_{m} M_{m}^{\dagger} M_{m} = \mathbb{I},
\end{equation}

meaning that probabilities must sum to one, i.e, $|\psi\rangle$ is normalized:

\begin{equation} 
\sum_{m} Pr(o_m) = \sum_{m} \langle \psi | M^{\dagger}_{m} M_{m} |\psi \rangle = 1.
\end{equation} 

## Positive Operator-Valued Measure (POVM)

The formalism of POVM is a consequence of the general description for measurements.

A POVM is a set of operators $\{E_m\}$ with the following two properties:

1. $\langle \psi |E_m| \psi \rangle \geq 0, \forall |\psi\rangle$. Meaning that $E_m$ is a positive semidefinite operator.

2. $\sum_m E_m = \mathbb{I}$. Completeness relation.

In this formalism, the probability of obtaining outcome $m$ after applying a POVM is given by:

$$Pr(m) := \langle \psi | E_m | \psi \rangle.$$

## Projective measurement (a.k.a von Neumann measurement) and Born's rule

A projective measurement is a special case of a POVM where all POVM elements $E_m$ are the same as the measurement operators $M_m$:

$$E_m = \Pi_m^{\dagger} \Pi_m = \Pi^2_m = \Pi_m = M_m.$$

This is true because all orthogonal projection operators are hermitian ($\Pi^{\dagger} = \Pi$) and idempotent ($\Pi^2 = \Pi$).

Also, recall that projector operators are positive operators.

- **First case: Discrete Spectrum.**

Suppose the state of a quantum system is prepared in the basis of the observable $\hat{\mathcal{O}}$ one wants to measure, i.e, $|\psi\rangle \doteq \sum_{j=1}^d c_j|o_j\rangle$. A **projective measurement** (a.k.a **von Neumann measurement**) entails the following conditional probability for obtaining the eigenvalue $o_j$ of $\hat{\mathcal{O}}$:

\begin{align} 
Pr(o_j|\psi\rangle) &= \langle \psi | \Pi^{\dagger}_{o_j} \Pi_{o_j} |\psi \rangle \\ 
&= \langle \psi | \Pi_{o_j}^2 |\psi \rangle\\
&= \langle \psi| (|o_j\rangle \langle o_j|o_k\rangle \langle o_k|)| \psi \rangle\\
&=\delta_{jk}\langle\psi|o_j\rangle \langle o_k |\psi\rangle \\ 
&= \langle\psi|o_j\rangle \langle o_j |\psi\rangle \\ 
&= \langle \psi |o_j\rangle \langle \psi| o_j \rangle^*\\
&= |\langle \psi|o_j \rangle|^2=|\langle o_j|\psi \rangle|^2\\
&= \left|\langle o_j|\sum_{k=1}^{d} c_k |o_k\rangle\right|^2\\
&=  \left|\sum_{k=1}^{d} c_k \langle o_j|o_k\rangle\right|^2\\
&=\left|\sum_{k=1}^{d} c_k \delta_{jk} \right|^2 \\
&= |c_j|^2.
\end{align} 

This result is also known as **Born's rule**:

\begin{equation}
Pr(o_j|\psi\rangle) \doteq |\langle o_j |\psi \rangle|^2 = |\langle \psi |o_j \rangle|^2= |c_j|^2.
\end{equation}

Note that $Pr(o_j|\psi\rangle)=1$ if and only if $|\psi\rangle = e^{i\phi_0} |o_j\rangle$, where $|o_j\rangle$ is an eigenstate of $o_j$.

The normalization condition entails:

$$
||\psi||^2 =
\langle \psi |\psi\rangle =
|\psi\rangle^{\dagger} |\psi\rangle  =
\Bigg(\sum_{j=1}^{d}c_j|o_j\rangle \Bigg)^{\dagger}\Bigg(\sum_{k=1}^{d} c_k |o_k\rangle \Bigg) =
\sum_{j,k=1}^{d} c_j^* c_k \langle o_j|o_k\rangle= \sum_{j,k=1}^{d} c_j^* c_k \delta_{jk} = 
\sum_{j=1}^{d}|c_j|^2 =
\sum_{j=1}^{d}Pr(o_j|\psi\rangle)=1.
$$

If the system has evolved from the initial state $|\psi(0)\rangle$ to the final state $|\psi(t)\rangle$, then:

\begin{equation}
Pr(o_j|\psi_t\rangle) = |\langle o_j | \hat{U}(t) |\psi(0)\rangle|^2 = |\langle o_j |\psi(t)\rangle|^2.
\end{equation}

- **Second case: Continuous Spectrum.**

Consider the state of a quantum system prepared in the basis of the observable $\hat{\mathcal{Q}}$ one wants to measure, i.e, $|\psi\rangle \doteq \int \psi_{q} |q\rangle d_{q}$. A **projective measurement** (a.k.a **von Neumann measurement**) entails the following conditional probability for obtaining the eigenvalue $q$ of $\hat{\mathcal{Q}}$:

\begin{align}
Pr(q|\psi\rangle) &= |\langle q|\psi \rangle|^2\\
&= \left|\langle q|\int \psi_{q'} |q'\rangle d_{q'}\rangle\right|^2\\
&=  \left|\int \psi_{q'} \langle q|q'\rangle d_{q'}\right|^2\\
&=\left|\int \psi_{q'} \delta(q-q') d_{q'} \right|^2 \\
&= |\psi_{q}|^2.    
\end{align}

## Expectation value

The expected value of a von Neumann observable $\hat{\mathcal{O}}$ (e.g., the Hamiltonian) of a quantum system prepared in the state $|\psi\rangle$ is the sum of the eigenvalues of such observable weighted by their corresponding Born rule probabilities: 

\begin{align}
\langle \hat{\mathcal{O}}\rangle &\doteq \sum_{j=1}^{d=2^n} o_j Pr(o_j|\psi \rangle)\\
&=\sum_{j=1}^{d=2^n} o_j|\langle o_j |\psi \rangle|^2 \\
&=\sum_{j=1}^{d=2^n} o_j  \langle \psi |o_j \rangle \langle \psi |o_j \rangle^* \\
&=\sum_{j=1}^{d=2^n} o_j  \langle \psi |o_j \rangle \langle o_j|\psi \rangle \\
&= \langle \psi | \left(\sum_{j=1}^{d=2^n} o_j |o_j \rangle \langle o_j| \right) |\psi \rangle\\
&= \langle \psi | \left(\sum_{j=1}^{d=2^n} o_j P_{o_j} \right) |\psi \rangle\\
&=\langle \psi |\hat{\mathcal{O}}|\psi\rangle \\
&= \langle 0^{\otimes n}|\hat{U}^{\dagger} \hat{\mathcal{O}} \hat{U}|0^{\otimes n}\rangle \in \mathbb{R}.
\end{align}

Where $\hat{U}$ denotes the unitary evolution operator that evolves the initial state $|0^{\otimes n}\rangle$ to the final state $|\psi\rangle$.

Note: In the quantum circuit approach, the evolution operator $\hat{U}$ is represented by a sequence of gates.


Understanding the last expression:

1. Every quantum computer of $n$ qubits start at the initial state denoted by $|0^{\otimes n} \rangle := |0 \rangle \otimes \cdots \otimes |0 \rangle$.

2. One can then write the evolved state as:

\begin{align}
|\psi \rangle &= \hat{U} |0^{\otimes n} \rangle, \\
\langle \psi | &= |\psi \rangle^{\dagger}= (\hat{U} |0^{\otimes n} \rangle)^{\dagger} = |0^{\otimes n} \rangle^{\dagger}\hat{U}^{\dagger} = \langle 0^{\otimes n}|\hat{U}^{\dagger}.
\end{align}



## Examples

If the state $|\psi\rangle$ of the quantum system is an eigenstate of the observable being measured, the measurement outcome is `deterministic`. However, when $|\psi\rangle$ is not an eigenstate of the observable, then the measurement outcome is `probabilistic`. The latter is a consequence of Heisenberg's uncertainty principle for conjugate observables that do not commute, i.e, $[\hat{A},\hat{B}] \ne \mathbb{O}$.

Examples:

- Consider a quantum system in the state $|\psi\rangle=|0\rangle$.
  - A measurement in the $\hat{Z}$ basis is deterministic: $Pr(|0\rangle)=|\langle 0|\psi \rangle|^2=|\langle 0|0 \rangle|^2=1=100\%$.
  - A measurement in the $\hat{X}$ basis is probabilistic: $Pr(|+\rangle)=|\langle +|\psi \rangle|^2=|\langle 0|H|0 \rangle|^2=1/2=50\%$.
  
</br>

- Consider a quantum system in the state $|\psi\rangle=|+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$.
  - A measurement in the $\hat{Z}$ basis is probabilistic: $Pr(|0\rangle)=|\langle 0|\psi \rangle|^2=|\langle 0|+ \rangle|^2=1/2=50\%$.
  - A measurement in the $\hat{X}$ basis is deterministic: $Pr(|+\rangle)=|\langle +|\psi \rangle|^2=|\langle +|+ \rangle|^2=1=100\%$.
